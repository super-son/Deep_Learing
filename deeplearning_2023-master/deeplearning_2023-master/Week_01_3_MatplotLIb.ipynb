## Matplotlib

Matplotlib는 파이썬의 대표적인 데이터 시각화 라이브러리 중 하나로, 다양한 형식의 그래프와 차트를 생성하고 시각화하는데 사용됩니다.  
데이터의 분포, 관계, 패턴 등을 시각화하여 데이터를 이해하고 표현하는데 유용한 도구입니다.

```python
# 라이브러리 불러오기
import matplotlib.pyplot as plt
```

## 기본적인 사용법(선그래프)
선 그래프(Line Chart)는 데이터의 변화를 시간이나 다른 연속적인 변수에 따라 선으로 연결하여 시각화하는 그래프 유형입니다.  
주로 시계열 데이터나 연속적인 데이터의 변화를 표현할 때 사용됩니다.  
각 데이터 포인트는 점으로 표시되며, 이 점들을 선으로 연결하여 데이터의 추세나 변화를 보여줍니다.

```python
import matplotlib.pyplot as plt

# 데이터 생성
x = [1, 2, 3, 4, 5]
y = [10, 25, 18, 30, 15]

# 선 그래프 생성
plt.plot(x, y)
```

## 기본적인 사용법 (그래프 튜닝)
Matplotlib는 그래프의 많은 요소를 사용자 정의할 수 있는 기능을 제공합니다.  
사용자의 요구에 따라 그래프를 자세하게 맞춤할 수 있습니다.

```python
# 데이터 생성
x = [1, 2, 3, 4, 5]
y = [10, 25, 18, 30, 15]

# 선 그래프 생성
plt.plot(x, y)

# 축 레이블과 타이틀 추가
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Line Chart')

# 그래프 표시
plt.show()
```

## 산점도
산점도(Scatter Plot)는 데이터의 두 변수 간의 관계를 나타내기 위해 데이터 포인트를 점으로 표시하는 그래프 유형입니다.  
주로 연속적인 변수나 두 변수 간의 상관관계를 시각화할 때 사용됩니다.  
각각의 데이터 포인트는 x축과 y축에 대응되며, 데이터의 분포와 패턴을 파악하기에 유용한 도구입니다.

```python
import matplotlib.pyplot as plt

# 데이터 생성
x = [1, 2, 3, 4, 5]
y = [10, 25, 18, 30, 15]

# 산점도 생성
plt.scatter(x, y, color='r', marker='o', label='Data Points')

# 축 레이블과 타이틀 추가
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Scatter Plot')

# 범례 추가
plt.legend()

# 그래프 표시
plt.show()
```

## 히스토그램

히스토그램(Histogram)은 연속적인 데이터의 분포를 시각화하는 그래프 유형입니다.  
데이터의 빈도(데이터가 나타나는 횟수)를 나타내며, 각 데이터의 범위를 구간으로 나누어 그 범위 내에 속하는 데이터의 개수를 표시합니다.  
히스토그램을 통해 데이터의 분포, 중심 경향성, 이상치 등을 파악할 수 있습니다.

```python
import matplotlib.pyplot as plt

# 데이터 생성
data = [15, 11, 35, 12, 45, 50, 5, 80, 50]

# 히스토그램 생성
plt.hist(data, bins=4, color='b', edgecolor='black', alpha=0.7)

# 축 레이블과 타이틀 추가
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# 그래프 표시
plt.show()
```
